```python
import pandas 

cabecalho = ['Alunos', 'Média']
caderneta = pandas.DataFrame(columns=cabecalho)

quantidade_questoes = int(input(f'Quantas questões tem a prova?\n'))
quantidade_alunos = int(input(f'Quantos alunos fizeram a prova?\n'))

questao = 1
for i in range(quantidade_questoes):
  caderneta['Questão'+str(questao)] = ''
  questao = questao + 1

print('\nDADOS SOBRE O GABARITO\n')
dados_gabarito = []
dados_gabarito.append('GABARITO')
dados_gabarito.append('10')
for a in range(len(caderneta.columns) - 2):
  dados_gabarito.append(input(f'Qual a resposta da questão {a+1}?\n').upper())
caderneta.loc[len(caderneta.index)] = dados_gabarito

for j in range(quantidade_alunos):
  dados_prova = []
  dados_prova.append(str(input(f'Qual o nome do aluno {j+1}?\n')).upper())
  dados_prova.append('') #Média
  for z in range(quantidade_questoes):
    dados_prova.append(input(f'Qual alternativa o aluno marcou na {z+1}ª questão?\n').upper())
  caderneta.loc[len(caderneta.index)] = dados_prova

for i in range(caderneta.shape[0]):
  nota_aluno = 0
  for j in range(caderneta.shape[1]):
    if j==1 or j==0:
      pass
    else:
      if caderneta.iloc[i,j] == caderneta.iloc[0,j]:
        nota_aluno = nota_aluno + 1
    caderneta.iloc[i,1] = nota_aluno

caderneta.to_excel('notas_dos_alunos.xlsx', index=False)

caderneta
```
